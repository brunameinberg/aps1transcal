```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Dados

```python

k = 240 
t_b = 100 + 273.15
t_inf = 25 + 273.15
teta_b = t_b-t_inf

#retangular
w = 0.24
t = 0.02
l = 0.175

#coef_conv
h = 25

p = 2*(w+t)
a_t = w*t
a_l = 2*l*w + 2*l*t


```

### 1) 
Qual a temperatura na extremidade da aleta? Apresente um gráfico da distribuição de temperatura ($${T}_{analítica}$$) 
ao longo do comprimento da aleta. Use legenda, identifique os eixos e as unidades usadas.

```python
m = sqrt((h*p)/(k*a_t))
alpha = (h/(m*k))

def dis_temp (lista_x):
   temperaturas = []
   for x in lista_x:
        t = teta_b*((cosh(m*(l-x))+alpha*(sinh(m*(l-x))))/(cosh(m*l)+alpha*sinh(m*l)))+t_inf
        temperaturas.append(t)
   return temperaturas

comprimentos = np.arange(0,l,0.01)

print("Temperatura na extremidade da aleta: {:.3f} K".format(dis_temp(comprimentos)[-1]))

plt.figure(figsize=(8, 6))
plt.plot(comprimentos,dis_temp(comprimentos))
plt.title("Temperatura (K) x Comprimento (m)")
plt.xlabel("Comprimento (m)")
plt.ylabel("Temperatura (K)")
plt.grid(True)

plt.show()
```

### 2) Taxa de transferência de calor

```python
M = sqrt(h*p*k*a_t*teta_b)
rel = (sinh(m*l)+alpha*cosh(m*l))/(cosh(m*l)+alpha*sinh(m*l))
qa = M*rel
print("Taxa de transferência de calor: {:.3f} W".format(qa))
```

### 3) Eficiência da aleta

```python
n = qa/(h*(a_l+ a_t)*teta_b)
print("Eficiencia: {:.3f}".format(n))
```

### 4) Efetividade da aleta

```python
efe = qa/(h*a_t*teta_b)
print("Efetividade: {:.3f}".format(efe))
```

### 5) Taxa de transferência de calor na aleta infinita


```python
qa_inf = teta_b*((h*p*k*a_t)**0.5)
print("Taxa de transferência de calor na aleta infinita: {:.3f}".format(qa_inf))
```
