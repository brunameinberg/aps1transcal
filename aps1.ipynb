```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

k = 240 
t_b = 100 + 273.15
t_inf = 25 + 273.15
teta_b = t_b-t_inf

#retangular
w = 0.24
t = 0.02
l = 0.175

#coef_conv
h = 25

p = 2*(w+t)
a_t = w*t

```

### 1)

```python
m = sqrt((h*p)/(k*a_t))
alpha = (h/(m*k))

def dis_temp (lista_x):
   temperaturas = []
   for x in lista_x:
        t = teta_b*((cosh(m*(l-x))+alpha*(sinh(m*(l-x))))/(cosh(m*l)+alpha*sinh(m*l)))+t_inf
        temperaturas.append(t)
   return temperaturas

comprimentos = np.arange(0,l,0.01)
print(dis_temp(comprimentos)[-1])
plt.plot(comprimentos,dis_temp(comprimentos))
plt.show() 
```
